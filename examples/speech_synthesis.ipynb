```python
import IPython
from pororo import Pororo
```

```python
tts = Pororo("tts", lang="multi")
```

```python
# Typical TTS
wave = tts("  !", lang="ko")
IPython.display.Audio(wave, rate=22050)
```

```python
# Cross-lingual Voice Style Transfer
wave = tts("   .", lang="ko", speaker="en")
IPython.display.Audio(wave, rate=22050)
```

```python
# Cross-lingual Voice Style Transfer
wave = tts("   .", lang="ko", speaker="en-18,ko")
IPython.display.Audio(wave, rate=22050)
```
